**Top 10 movies with 40+ ratings:**

```python
# count the rating for each item/film
ratingCount = ratingData.groupby("itemId")["rating"].count()
ratingCount
```

```python
# a series of indexes which are movies with more than 40 ratings
mostRatedFilmsIndex = ratingCount[ratingCount > 40].index
mostRatedFilmsIndex
```

```python
# table for the top ten films
mostRatedFilms = ratingData[ratingData['itemId'].isin(mostRatedFilmsIndex)] # all films with over 40 ratings 
topTenFilms = (
        mostRatedFilms
        .groupby("itemId")["rating"].mean() # calculate the mean of each movies ratings
        .reset_index() # convert to data frame
        .sort_values("rating", ascending=False) # top = the highest rated
        .head(10) # top ten
) 
topTenFilms
```

```python
#calculate % of male students 
maleStudents = menOccupation[menOccupation["occupation"] == "student"] # students
procentOfMaleStudents = menOccupation["count"].sum() / maleStudents["count"] # calculate the procent

print(f"Out of all males {procentOfMaleStudents.iloc[0]:.2f} % are students.")
```

**Combine u.data and u.item so that we have a table with Movie Title and Rating:**

```python
# how does the film data look again ?
filmData.head(2)
```

```python
# how does the rating data look again ?
ratingData.sort_values("rating").head(5)
```

```python
# merge the two tables to get all info in one
ratingAndFilmData = pd.merge(ratingData, filmData, left_on="itemId", right_on="movie id") # inner join, on= ratingData.itemId : filmData.movie id
ratingAndFilmData
```

```python
# show more readable
ratingAndFilmData[["movie id", "movie title", "rating"]] # print new joined table but we choose to show only 3 columns 
```

**Combine the previously made table with userdata to put it all together:** 

```python
# reminder of how userData looks like
userData.head(3)
```

```python
# reminder of how ratingAndFilmData looks like
ratingAndFilmData.head(3)
```

```python
# merge them
fullTable = pd.merge(ratingAndFilmData, userData, left_on="userId", right_on="user id") # how=inner, on=ratingAndFilmData.userId : userData.user id
fullTable.head(3)
```

```python
# display all columns except genres, and sort values according to user id
fullTable[["movie id", "movie title", "rating", "user id", "occupation", "gender", "age"]].sort_values("user id")
```

```python
# show all for one user
fullTable[fullTable["user id"] == 1][["movie id", "movie title", "rating", "user id", "occupation", "gender", "age"]]
```

user#1 really likes movies, 272 watched!!^^

**Average rating per user**

```python
meanRatingPerUser = (
        fullTable
        .groupby("user id") #group users
        ["rating"] # we want to look at rating
        .mean() # calc the mean of rating per user
        .reset_index() # make into data frame
        .sort_values("rating", ascending=False) # sort by rating (descending)
)
meanRatingPerUser.rename(columns={"rating": "mean rating"}, inplace=True) # change column name to represent mean 
meanRatingPerUser
```

**more investigation:**

```python
# top 5 positive mean rating user
topPositiveUsers = meanRatingPerUser.head(5)
topPositiveUsers
```

```python
# merge our full table with positive users to get more info in one table
topFive = pd.merge(fullTable, topPositiveUsers, left_on="userId", right_on="user id")
noDupTopFive = topFive.drop_duplicates(subset="userId") # drop duplicate rows of "user id"
noDupTopFive[["userId", "occupation", "age", "gender", "mean rating"]].sort_values("mean rating", ascending=False) # show relevant columns
```

```python
# top 5 negative mean rating user
topNegativeUsers = meanRatingPerUser.tail(5)
topNegativeUsers
```

```python
# merge our full table with negative users
botFive = pd.merge(fullTable, topNegativeUsers, left_on="userId", right_on="user id")
noDupBotFive = botFive.drop_duplicates(subset="userId") #show only one of each userId
noDupBotFive[["userId", "occupation", "age", "gender", "mean rating"]].sort_values("mean rating")  # show relevant columns
```

21+ aged people are more critical than younger people, it seems. 

Students watch alot of movies.

```python
# merge whole table
duplicateMegaTable = pd.merge(fullTable, meanRatingPerUser, left_on="userId", right_on="user id")
megaTable = duplicateMegaTable.drop_duplicates("userId") #drop duplicate rows
megaTable[["userId", "occupation", "age", "gender", "mean rating"]].sort_values("mean rating")
```
